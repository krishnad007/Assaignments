```python
import matplotlib.pyplot as mat
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm
import pandas as pan
import numpy as np
```

```python
iris=datasets.load_iris()
```

```python
print(iris.target)
```

```python
print(iris.target_names)
```

```python
X=pan.DataFrame(iris.data)
```

```python

X.columns=['Sepal_length','Sepal_width','Petal_length','Petal_width']
Y=pan.DataFrame(iris.target)
Y.columns=['Target']

mat.figure(figsize=(14,14))

colormap=np.array(['red','black','blue'])

mat.subplot(2,1,1)

mat.scatter(X.Petal_length,X.Petal_width,c=colormap[Y.Target],s=40)

mat.title('Petal')
mat.xlabel('Petal Length')
mat.ylabel('Petal Width')

mat.subplot(2,1,2)

mat.scatter(X.Sepal_length,X.Sepal_width,c=colormap[Y.Target],s=40)

mat.title('Sepal')
```

```python
from sklearn.cluster import AgglomerativeClustering
Hcluster=AgglomerativeClustering(n_clusters=3)
Hcluster.fit(X)
pred_Y=Hcluster.labels_
pred_Y
```

```python
mat.figure(figsize=(10,10))

mat.subplot(2,1,1)

mat.scatter(X.Petal_length,X.Petal_width,c=colormap[Y.Target],s=40)

mat.title('Real classification')

mat.subplot(212)

mat.scatter(X.Petal_length,X.Petal_width,c=colormap[pred_Y],s=40)

mat.title('KMean')
```
